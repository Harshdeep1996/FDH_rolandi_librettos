```python
import re
from bs4 import BeautifulSoup

#connect to a URL
url = 'http://dl.cini.it/collections/show/1120'
website = urllib.request.urlopen(url)

#read html code
html = website.read()
soup = BeautifulSoup(html)
links = soup.findAll("a")

#To get href part alone
links[1]
```

```python
numbers = []

for i in range(5, len(links)):
    numbers.append(int(str(links[i].get('href')).split('/')[-1]))
```
